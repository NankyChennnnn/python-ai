# 逻辑回归

### 1 分类，公式前半部分

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体为苹果系统常用字体
plt.rcParams['font.sans-serif'] = ['Heiti SC', 'PingFang SC', 'Songti SC', 'Arial Unicode MS', 'SimHei']  # 依次尝试这些字体
plt.rcParams['axes.unicode_minus'] = False  # 处理负号显示问题

# 生成p的取值，从0.01到1，防止log(0)产生无限大
p = np.linspace(0.01, 1, 100)
y = -np.log(p)

plt.figure(figsize=(6,4))
plt.plot(p, y, label='-log(p)')
plt.xlabel('p')
plt.ylabel('-log(p)')
plt.title('0-1的 -log(p) 曲线')
plt.grid(True)
plt.legend()
plt.show()
```

### 0 分类，公式后半部分

```python
# 绘制 0-1 区间的 -log(1-p) 曲线
p = np.linspace(0.01, 0.99, 100)  # 避免 1-p 等于0，防止log(0)
y = -np.log(1 - p)

plt.figure(figsize=(6,4))
plt.plot(p, y, label='-log(1-p)', color='orange')
plt.xlabel('p')
plt.ylabel('-log(1-p)')
plt.title('0-1的 -log(1-p) 曲线')
plt.grid(True)
plt.legend()
plt.show()
```
