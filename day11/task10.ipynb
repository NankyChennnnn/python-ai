### 练习series和df的索引操作

```python
import pandas as pd
import numpy as np

# 用Series随机生成一个对象
s = pd.Series(np.random.randn(5), index=list('abcde'))
print(s)
```

```python
# 以元素 'b' 的索引和切片为例
print("使用标签索引's': s['b'] =", s['b'])
print("使用位置索引's': s.iloc[1] =", s.iloc[1])
print("使用标签切片's': s['b':'d'] =\n", s['b':'d'])
print("使用布尔索引's': s[s > 0] =\n", s[s > 0])
```

```python
# 用DataFrame随机生成一个对象
df = pd.DataFrame(np.random.randn(5, 4), index=list('abcde'), columns=list('WXYZ'))
print(df)
```

```python
# 案例：演示DataFrame的常用索引操作

# 通过标签选取某一行
print("df.loc['a'] (标签选取行):\n", df.loc['a'])

# 通过位置选取某一行
print("\ndf.iloc[0] (位置选取行):\n", df.iloc[0])

# 通过标签选取某一列
print("\ndf['W'] (选取列):\n", df['W'])

# 通过标签切片选取多行多列
print("\ndf.loc['b':'d', 'X':'Z'] (行、列标签切片选取):\n", df.loc['b':'d', 'X':'Z'])
```
