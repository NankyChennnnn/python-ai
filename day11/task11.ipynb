### 练习缺失值处理

```python
import numpy as np
import pandas as pd

# 随机生成一个5x6的DataFrame
df = pd.DataFrame(np.random.randn(5, 6))

# 随机选择至少4个位置设为NaN
nan_indices = np.random.choice(df.size, 4, replace=False)
for idx in nan_indices:
    row, col = divmod(idx, df.shape[1])
    df.iat[row, col] = np.nan

print(df)
```

```python
# 按行统计缺失率
row_nan_rate = df.isna().mean(axis=1)
print("按行统计缺失率：")
print(row_nan_rate)

# 按列统计缺失率
col_nan_rate = df.isna().mean(axis=0)
print("\n按列统计缺失率：")
print(col_nan_rate)
```

```python
# 按行删除缺失数据所在行
df_dropna_row = df.dropna(axis=0)
print("按行删除缺失数据所在行：")
print(df_dropna_row)

# 按列删除缺失数据所在列
df_dropna_col = df.dropna(axis=1)
print("\n按列删除缺失数据所在列：")
print(df_dropna_col)
```

```python
# 计算每列均值
col_means = df.mean()
# 用均值填补nan数据
df_filled = df.fillna(col_means)
print("用每列均值填补缺失值后的DataFrame：")
print(df_filled)
```
