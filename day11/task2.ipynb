### 掌握numpy的数据类型，练习数组和数的运算，数组和数组的运算，理解数组的轴

#### 数据类型

```python
import numpy as np

# 获取所有numpy支持的数据类型
all_dtypes = np.sctypeDict

# 只保留不重复的dtype对应名称
unique_dtypes = set(np.sctypeDict.values())

# 打印所有可用的数据类型（dtype方式）
print("NumPy支持的数据类型（dtype）：")
for dtype in sorted(unique_dtypes, key=lambda x: str(x)):
    print(dtype)
```

#### 数组、数、轴

```python
# 创建一个 (3,4) 的 ndarray
arr = np.arange(12).reshape(3, 4)
print("原数组：\n", arr)

# 数组与数的运算: 每个元素+10
arr_plus_10 = arr + 10
print("每个元素加10后的数组：\n", arr_plus_10)

# 数组与数组的运算: 与自身相乘
arr_squared = arr * arr
print("数组与自身相乘：\n", arr_squared)

# 理解数组的轴
print("数组 shape:", arr.shape)
print("按 axis=0 求和：", np.sum(arr, axis=0))  # 列和
print("按 axis=1 求和：", np.sum(arr, axis=1))  # 行和
```
