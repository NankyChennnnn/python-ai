### 完成nan和inf的使用,与上课一致

```python
import numpy as np

np.nan != np.nan
```

```python
arr = np.random.rand(3, 4)
# 随机选择几个元素设为nan
num_nan = np.random.randint(1, 4)  # 随机设置1到3个nan
nan_indices = np.random.choice(arr.size, num_nan, replace=False)
arr.flat[nan_indices] = np.nan
print(arr)
```

```python
np.isnan(arr)
```

```python
arr != arr
```

```python
arr = np.array([1, 0])
result = arr[0] / arr[1]
print(result)
```

```python
np.inf > -np.inf
```
