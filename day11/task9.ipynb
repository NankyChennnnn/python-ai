### 练习字典转series与df

```python
import pandas as pd

# dict transform to series
dict_obj = {'Tom': 90, 'Bob': 85, 'Lucy': 88}
ser_from_dict = pd.Series(dict_obj)
ser_from_dict
```

```python
import numpy as np

dict_data = {'A': 1,
             'B': pd.Timestamp('20190926'),
             'C': pd.Series(1, index=list(range(4)),dtype='float32'),
             'D': np.array([3] * 4,dtype='int32'),
             'E': ["Python","Java","C++","C"],
             'F': 'wangdao' }

# dict transform to dataframe
df_from_dict = pd.DataFrame(dict_data)
df_from_dict.index=['a','b','c','d']
df_from_dict
```
