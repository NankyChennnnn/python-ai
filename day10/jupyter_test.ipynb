# 开始绘图

## 折线图

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 6, 5]

plt.plot(x, y, marker='o')
plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Hiragino Sans GB', 'Heiti SC', 'Arial Unicode MS', 'Helvetica', 'Arial']  # MacOS常用中文和英文字体
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号
plt.title('折线图示例')
plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = range(1, 8)  # x轴的位置
y = [17, 17, 18, 15, 11, 11, 13]
# 设置中文字体为系统支持的常用字体，避免中文乱码
plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Hiragino Sans GB', 'Heiti SC', 'Arial Unicode MS', 'Helvetica', 'Arial']  
# 设置坐标轴负号可以正常显示
plt.rcParams['axes.unicode_minus'] = False  
# 设置图表标题
plt.title('每周销售数量')  
# 设置x轴标签
plt.xlabel('星期')  
# 设置y轴标签
plt.ylabel('销售量')  
# 显示网格线，便于观察数据
plt.grid(True)

plt.plot(x, y, color='red', alpha=0.5, linestyle='--', linewidth=3, marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = range(2, 26, 2)  # x轴的位置
y = [random.randint(15, 30) for i in x]  # y轴的温度

plt.figure(figsize=(16, 8), dpi=80)
plt.plot(x, y, marker='o')

plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Hiragino Sans GB', 'Heiti SC', 'Arial Unicode MS', 'Helvetica', 'Arial']  # 解决中文乱码
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

plt.title('一天温度变化折线图')
plt.xlabel('小时')
plt.ylabel('温度(℃)')
plt.xticks(x, [f'{i}时' for i in x], rotation=45)
plt.yticks(range(15, 31), [f'{i}℃' for i in range(15, 31)])
plt.grid(True)
plt.tight_layout()

# plt.savefig('line_chart.png')  # 保存为png格式
plt.savefig('line_chart.svg')  # 保存为svg格式
plt.show()  # show 之后，图像对象会被释放
```
